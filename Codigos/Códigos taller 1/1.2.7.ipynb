```python
import sympy as sp
```

```python
# se definen los vectores

a = sp.Matrix([1,2,3])
b = sp.Matrix([4,5,6])
c = sp.Matrix([3,2,1])
d = sp.Matrix([6,5,4])

#sumas

print('6(a)sumas ')
suma1 = a+b+c+d
suma2 = a+b-c-d
suma3 = a-b-c-d
suma4 = -a +b-c+d
print(suma1, suma2, suma3, suma4)

#se define la base canonica

print('6(b) angulo entre los vectores y la base')
e1 = sp.Matrix([1,0,0])
e2 = sp.Matrix([0,1,0])
e3 = sp.Matrix([0,0,1])
# se define la ecuacion para hallar el angulo cos(alpha)=(a.b)/(|a||b|) de manera general
def angulos(u,v):
  return sp.acos(u.dot(v)/(u.norm()*v.norm()))

angulos_a = [angulos(a,e1), angulos(a,e2), angulos(a,e3)]
angulos_b = [angulos(b,e1), angulos(b,e2), angulos(b,e3)]
angulos_c = [angulos(c,e1), angulos(c,e2), angulos(c,e3)]
angulos_d = [angulos(d,e1), angulos(d,e2), angulos(d,e3)]
print(angulos_a, angulos_b, angulos_c, angulos_d)

# hallamos la norma con .norm()

print('6(c) magnitud de los vectores a b c d ')
print(a.norm(), b.norm(), c.norm(), d.norm())

#se utiliza de nueva la función angulos para hallar angulos

print('6(d) angulo entre a y b y entre c y d')
angulo_ab = [angulos(a,b)]
angulo_cd = [angulos(c,d)]
print(angulo_ab, angulo_cd)

# proyección = ((a.b)/(|b|²))*b

print('6(e) proyección de a sobre b')
proyeccion_ab = (a.dot(b)/b.dot(b))*b
print(proyeccion_ab)

# verificar si los vectores a, b, c y d son coplanares
print('6(f) verificar si los vectores a, b, c y d son coplanares')
print(a.cross(c).dot(d))
print(a.cross(b).dot(c))
print(a.cross(b).dot(d))
print(b.cross(c).dot(d))
print('son coplanares')


print('6(g) encontrar (a+b).(c+d)')
print((a+b).dot(c+d))

print('6(h) hallar los productos axb, bxc, cxd y los angulos que forman con d')
print(a.cross(b),[angulos(a.cross(b),c)])
print(b.cross(c),[angulos(b.cross(c),d)])
print(c.cross(d),[angulos(c.cross(d),a)])

print('6(i)')
print(c.dot(a.cross(b)))
```
