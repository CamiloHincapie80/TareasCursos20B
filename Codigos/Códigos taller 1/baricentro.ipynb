```python
import sympy as sp

#entrada de los puntos

try:
  x1 = float(input("Ingrese la coordenada x del primer punto: "))
  y1 = float(input("Ingrese la coordenada y del primer punto: "))
  a = sp.Point(x1, y1)

  x2 = float(input("Ingrese la coordenada x del segundo punto: "))
  y2 = float(input("Ingrese la coordenada y del segundo punto: "))
  b = sp.Point(x2, y2)

  x3 = float(input("Ingrese la coordenada x del tercer punto: "))
  y3 = float(input("Ingrese la coordenada y del tercer punto: "))
  c = sp.Point(x3, y3)

  # Verificar colinealidad
  if sp.Point.is_collinear(a, b, c):
    print("Los puntos son colineales y no forman un triángulo válido.")
  else:
    # Calcular el centroide
    # Centroide = (a + b + c) / 3
    centroide_x = (a.x + b.x + c.x) / 3
    centroide_y = (a.y + b.y + c.y) / 3
    centroide = sp.Point(centroide_x, centroide_y)
    print(f"\nEl centroide del triángulo es: {centroide}")

    # Calcular el circuncentro
    ox, oy = sp.symbols('ox oy')
    O = sp.Point(ox, oy)

    # Ecuación 1: |O - p1|^2 = |O - p2|^2
    # (ox - x1)^2 + (oy - y1)^2 = (ox - x2)^2 + (oy - y2)^2
    ecuacion1 = sp.Eq((ox - a.x)**2 + (oy - a.y)**2, (ox - b.x)**2 + (oy - b.y)**2)

    # Ecuación 2: |O - p1|^2 = |O - p3|^2
    # (ox - x1)^2 + (oy - y1)^2 = (ox - x3)^2 + (oy - y3)^2
    ecuacion2 = sp.Eq((ox - a.x)**2 + (oy - a.y)**2, (ox - c.x)**2 + (oy - c.y)**2)

    # Resolver el sistema de ecuaciones lineales para ox y oy
    solucion = sp.solve((ecuacion1, ecuacion2), (ox, oy))

    if solucion:
      circuncentro = sp.Point(solucion[ox], solucion[oy])
      print(f"El circuncentro del triángulo es: {circuncentro}")
    else:
        print("No se pudo calcular el circuncentro.")


except ValueError:
  print("Entrada inválida. Por favor, ingrese números para las coordenadas.")
except Exception as e:
    print(f"Ocurrió un error: {e}")
```
