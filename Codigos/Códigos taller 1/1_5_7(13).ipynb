```python
import sympy as sp

# variables
theta = sp.symbols('theta', real=True)

# parametrización circunferencia unidad
x = sp.cos(theta)
y = sp.sin(theta)

# campo evaluado sobre la curva
Fx = -y/(x**2 + y**2)
Fy =  x/(x**2 + y**2)

# dr/dtheta
dx = sp.diff(x, theta)
dy = sp.diff(y, theta)

# integrando F·(dr/dtheta)
integrand = sp.simplify(Fx*dx + Fy*dy)   # debe simplificar a 1

# trabajos en contra
W_a = -sp.integrate(integrand, (theta, 0, sp.pi))    # (a)
W_b = -sp.integrate(integrand, (theta, 0, -sp.pi))   # (b)


print("(a) Trabajo en contra (0→π, antihorario):", W_a)   # → -pi
print("(b) Trabajo en contra (0→-π, horario):", W_b)  # → pi
```
